## 1. Элементарные типы данных и операторы

Добро пожаловать в "блокнот Ipython" (он же "блокнот Jupyter")!

Перед вами штуковина, которая позволяет в одном файле совместить написание кода, его использование для анализа данных, и создание отчёта. 

Блокнот состоит из ячеек. Данная строка и две предыдущие находятся в одной ячейке. Текстовой.

А эта строка уже в следующей ячейке. Ячейки можно "потрогать" мышкой. Если, конечно, вы запускаете блокнот на своей машине.

По двойному клику в текстовой ячейке включается режим редактирования. По нажатию Ctrl+Enter или Shift+Enter - содержимое ячейки вычисляется. Попробуйте!

````python
# А это - ячейка для программного кода на языке Python. По нажатию Ctrl+Enter этот код выполняется.
# Знак диеза в начале строки служит для того, чтобы последующий текст не воспринимался как код.  А как комментарий.
# А вот следующая срока будет восприниматься. И запустится!
print('Hello World!')
print('Привет, Питон!')
# Команда print печатает под ячейкой то, что ей укажут. В данном случае - текст.
# Поехали! Жмите Ctrl+Enter
````

Успешно запустившаяся ячейка печатает слева от себя номер запуска (типа In [2]), а внизу от себя - текст или графику, если этого потребовал введённый в ней код.

Чтобы создать новую ячейку, нажмите кнопку 'Insert' в тулбаре сверху. 

Чтобы изменить тип ячейки, воспользуйтесь выпадающим списком ('Code' или 'Markdown')

Python можно использовать как большой калькулятор. Поэкспериментируйте в ячейке ниже с арифметическими действиями!

````python
print(2 * 2) 
print(2 ** 10) # ** - это оператор возведения в степень.
print(666 + 666 + 666 + 6 + 6 + 6)
````

Если не писать команду print, ячейка тоже выводит результаты операций. Но только одной - последней.

````python
2*2 # это число не напечатается
3*3 # и это тоже
4*4 # они все перетрутся последним числом
````

Кроме чисел, Python умеет работать с логическими значениями (True и False). 

При попытке совершать с ними арифметические действия, он конвертирует их в 1 и 0 соответственно.

Однако хорошим тоном является НЕ мешать логические значения с числовыми и вообще разные типы данных друг с другом.

````python
print(11 > 6)
# == - оператор равенства
print(2*2 == 5)
print(3.0 == 3)
print(True + True + True) # при попытке арифметических операций True превращается в 1
print(-50 * False) # при попытке арифметических операций False превращается в 0
````

## 2. Переменные, списки, циклы, условия

Как в любом нормальном языке программирования, в Питоне есть переменные - места в памяти, куда можно записывать любые данные.

Тип переменных определяется автоматически, никаким образом объявлять его не нужно.

````python
x = 5 * 5 # вычисляю 5*5 и записываю в x
y = 100 * x # использую этот x для вычисления y
print(y)
````

Переменные, рассчитанные в одной ячейке, доступны во всех ячейках. Пока вы не закроете блокнот, конечно.

Переменная $x$, созданная в прошлой ячейке, живёт в следующей. И во всех остальных тоже.

````python
print(x)
````

Несколько переменных можно объединять в <i>структуры данных</i>.

Самая популярная структура - список (aka массив). Это просто список значений через запятую, в квадратных скобках.

````python
lst = [1,2,3,4,5]
print(lst)
print(lst + [6,7,8]) # "сложение" списков - это присоединение одного списка к другому
````

Получить какой-то элемент списка можно по номеру с использованием квадратных скобок. Элементы нумеруюся С НУЛЯ.

````python
lst2 = ['The good', "The bad", "The ugly", 'And me!']
print(lst2[1])
print(lst2[0])
print(lst2[3])
````

Отрицательные индексы обозначают, что считать надо с конца (начиная с -1)

````python
print(lst2[-1])
````

Можно выгрызать сразу кусок списка, используя индексы от(включительно) и до(не включительно) через двоеточие.

Ещё раз: в Питоне первый индекс всегда включается, а последний - нет!

````python
print(lst2[1:3])
print(lst2[0:4])
````

То же самое можно делать с текстовыми строками: брать в них буквы по номеру.

````python
a = lst2[2]
print(a)
print(a[4:8] + ' me')
````

Оператор for...in позволяет пройтись по списку с начала до конца и что-то сделать с каждым элементом. Под это создаётся специальная переменная.

Внутренность оператора (что именно будем делать с элементами) отделяется отступом (tab либо четыре пробела).

В других языках внутренность чего-нибудь (оператора, функции и так далее) часто обозначается по-другому - например, фигурными скобками {}, как в языках семейства <b>C</b>. В Питоне всё это делается исключительно за счёт отступов.

````python
for element in lst2: # elmement - это то, как называется по очереди каждый элемент списка. Там могло быть и любое другое слово. 
    print(element)
print("Вот так-то!")
````

Можно использовать оператор for внутри одного списка для создания нового списка.

````python
numbers = [0,1,2,3,4,5]
squares = [x**2 for x in numbers]
print(squares)
````

Функция ```range(n)``` создаёт счётчик от ```0``` (включительно) до  ```n``` (не включительно).

````python
for i in range(0, 10):
    print('я буду прилежно учиться', i)
````

Похоже на for работает оператор if - проверяет, верно ли какое-то условие, и если верно, выполняет то, что внутри него.

Внутреннее содержание оператора точно так же выделяется отступом.

````python
if 5 < 6:
    print("Пять меньше чем шесть.")
elif 5 == 6:
    print("Пять равно шесть.")
else:
    print("Пять больше чем шесть.")

if 5 < 3:
    print("Пять меньше чем три.")
    print("Этот текст никогда не будет напечатан ;-(")
if 2 == 2.0: 
    print("Питон умеет сравнивать целые числа с дробными.")
    print("Кстати, в любом блоке кода может быть сколько угодно строк. Главное - правильно ставить отступы.")
````
