```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
plt.figure(figsize=(18, 6))
with plt.xkcd():
    matplotlib.rcParams.update({'font.size': 24})
    x = np.linspace(0, 10, 300)
    y = np.sqrt(x) + np.sin(x)
    plt.fill_between(x, y + np.sin(x)*0.3 + 0.2, y - np.sin(x) * 0.1 -0.1, color='cyan')
    plt.plot(x, y)
    plt.title('Это блокнот Jupyter c кодом на Python, выложенный на Github')
    plt.text(1.5, 0.5, 'А вы до сих пор пользуетесь Майкрософт Офисом?'.upper())
    plt.show()
```
